### lemniscate

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
alpha = 1
t = np.linspace(0, 2*np.pi, num=1000)
x = [alpha * np.sqrt(2)*np.cos(i) / (np.sin(i)**2+1) for i in t]
y = [alpha * np.sqrt(2)*np.cos(i)*np.sin(i) / (np.sin(i)**2+1) for i in t]
plt.plot(x, y)
#plt.show()
plt.savefig("lemniscate.png")
```
